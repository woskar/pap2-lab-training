```python
#Aufgabe 2: Dmpfung des Kreisels

#Import der Module
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

#Einlesen der Daten in folgender Reihenfolge:
t, w = np.loadtxt('tab1.txt', skiprows=1, usecols=(0,1), unpack=True)
#Zeit t in min
#Drefrequenz in rpm

t = t*60 #Umrechnen der Zeit in Sekunden 

# Fit mit einer Exponentialfunktion
def fit_func(x, A, d):
    return A*np.exp(-x*d)

#Fit durchfhren
popt, pcov = curve_fit(fit_func, t, w, maxfev=5000, p0=[700,0.0006], sigma=2)
perr = np.sqrt(np.diag(pcov))
x=np.arange(0, 750, 0.1)

#Gte des Fits
chisquare=np.sum((((fit_func(t,*popt))-w)/(2))**2)
dof=5 #degrees of freedom, Freiheitsgrad Anzahl Messwerte (7) - Fitparameter (2)
chisquare_red=chisquare/dof

from scipy.stats import chi2 
prob=round(1-chi2.cdf(chisquare,dof),2)*100 

#Plotten
plt.errorbar(t, w, yerr=2, linestyle='None', marker='x', label='$Messwerte$')
plt.plot(x, fit_func(x, *popt), label='$Fit: f(t)=Ae^{-\delta*t}$')
plt.xlabel('$Verstrichene\ Zeit\ in\ [s]$')
plt.ylabel('$Drehfrequenz\ \omega \ in\ [rpm]$')
plt.title(r'Diagramm 1: Dmpfung des Kreisels')
#plt.yscale('log')
#plt.grid(True)
plt.axis((0, 740, 400, 750))
plt.rcParams["figure.figsize"][0] = 8
plt.rcParams["figure.figsize"][1] = 5
plt.rcParams["font.family"]='serif'
plt.legend(title='Messwerte mit Fit', borderpad=0.5, borderaxespad=1, loc='best', shadow='true')

#Text im Plot
plt.text(38, 530, u'$A_{fit}=%.2f \pm%.2f$' %(popt[0], perr[0]), fontsize=10)
plt.text(38, 505, u'$\delta_{fit}=%.3f \pm%.3f\ [10^{-4}\ \\frac{1}{s} ]$' %(popt[1]*10000, perr[1]*10000), fontsize=10)
plt.text(38, 480, u'$\chi^{2}=%.2f$' %(chisquare), fontsize=10)
plt.text(38, 455, u'$\chi_{red}^{2}=%.2f $' %(chisquare_red), fontsize=10)
plt.text(38, 430, u'$P=%.2f $' %(prob) + '%', fontsize=10)


print('Werte in roher Form:')
print('Beste Werte aus Fit:')
print(popt)
print('Fehler der Werte aus Fit (Wurzel der Kovarianzmatrixwerte)')
print(perr)
print('Chiquadratsumme:')
print (chisquare)
print("Fitwahrscheinlichkeit="+str(prob)+"%")

#plt.savefig('v213_1.pdf',format='pdf')
```

```python
#Aufgabe 3: Przession

#Import der Module
#%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit

#Dmpfungskonstante aus Aufgabe 2: 
d = 0.0006531
dd = 0.0000042

#Einlesen der Daten in folgender Reihenfolge:
wf, tp = np.loadtxt('tab3.txt', skiprows=1, usecols=(0,1), unpack=True)
#wf Drefrequenz in rpm mit Fehler +/- 2
#tp Przessionsdauer in sec mit Fehler +/- 1

#Korrektur der wf durch Dmpfung
wfm = 0.5*wf*(1+np.exp(-d*tp))
dwfm = 0.5*((((1+np.exp(-d*tp))*2)**2+(wf*1*d*np.exp(-d*tp))**2+(wf*dd*tp*np.exp(-d*tp))**2)**0.5)

#Umrechnen der Drehfrequenz: w = 2pi*f, f=1/T und diese wurde gemessen in rpm
wfm = 2*np.pi*wfm/60
dwfm = 2*np.pi*dwfm/60


#Slicing der einzelnen Konfigurationen
wf1 = wfm[0:4] # 1 Masse in Abstand 15cm
dwf1 = dwfm[0:4]
tp1 = tp[0:4]
wf2 = wfm[4:8] # 1 Masse in Abstand 20cm
dwf2 = dwfm[4:8]
tp2 = tp[4:8]
wf3 = wfm[8:12] # 2 Massen in Abstand 15cm
dwf3 = dwfm[8:12]
tp3 = tp[8:12]
wf4 = wfm[12:16] # 2 Massen in Abstand 20cm
dwf4 = dwfm[12:16]
tp4 = tp[12:16]


# Fit mit Ursprungsgerade
def fit(x, m):
    return m*x
x=np.arange(0, 700, 0.1) #Punkte fr Darstellung des Fits

#Fit durchfhren
#Konfiguration 1
popt1, pcov1 = curve_fit(fit, wf1, tp1, maxfev=5000, sigma=dwf1)
perr1 = np.sqrt(np.diag(pcov1))

#Konfiguration 2
popt2, pcov2 = curve_fit(fit, wf2, tp2, maxfev=5000, sigma=dwf2)
perr2 = np.sqrt(np.diag(pcov2))

#Konfiguration 3
popt3, pcov3 = curve_fit(fit, wf3, tp3, maxfev=5000, sigma=dwf3)
perr3 = np.sqrt(np.diag(pcov3))

#Konfiguration 4
popt4, pcov4 = curve_fit(fit, wf4, tp4, maxfev=5000, sigma=dwf4)
perr4 = np.sqrt(np.diag(pcov4))


#Plotten
plt.errorbar(wf1, tp1, yerr=dwf1, linestyle='None', marker='x', label='$1: Eine\ Masse\ in\ 15\ cm$')
plt.errorbar(wf2, tp2, yerr=dwf2, linestyle='None', marker='x', label='$2: Eine\ Masse\ in\ 20\ cm$')
plt.errorbar(wf3, tp3, yerr=dwf3, linestyle='None', marker='x', label='$3: Zwei\ Massen\ in\ 15\ cm$')
plt.errorbar(wf4, tp4, yerr=dwf4, linestyle='None', marker='x', label='$4: Zwei\ Massen\ in\ 20\ cm$')

plt.plot(x, fit(x, *popt1), label='$ Fit\ 1: s_{1}=%.3f \pm%.3f$' %(popt1[0], perr1[0]))
plt.plot(x, fit(x, *popt2), label='$ Fit\ 2: s_{2}=%.3f \pm%.3f$' %(popt2[0], perr2[0]))
plt.plot(x, fit(x, *popt3), label='$ Fit\ 3: s_{3}=%.3f \pm%.3f$' %(popt3[0], perr3[0]))
plt.plot(x, fit(x, *popt4), label='$ Fit\ 4: s_{4}=%.3f \pm%.3f$' %(popt4[0], perr4[0]))

plt.xlabel('$Drehfrequenz\ \omega_{F} \ in\ [\\frac{1}{s}]$')
plt.ylabel('$Przessionsperiodendauer\ T_{p}\ in\ [s]$')

plt.title(r'Diagramm 2: Przessionsperiodendauer $T_{p}$ als Funktion der mittleren Drehfrequenz $\bar{\omega_{F}}$')
plt.grid(True)
plt.axis((0, 80, 0,130))
plt.rcParams["figure.figsize"][0] = 11
plt.rcParams["figure.figsize"][1] = 7
plt.rcParams["font.family"]='serif'
plt.legend(title='Messwerte mit linearer Fitfunktion $f(x)=s*x$ ', borderpad=1.0, borderaxespad=1, loc='best', shadow='true', ncol=2)

#plt.savefig('v213_2.pdf',format='pdf')
```

```python
#Aufgabe 4: Momentane Drehachse

#Import der Module
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

#Einlesen der Daten in folgender Reihenfolge:
wf, t = np.loadtxt('tab4.txt', skiprows=1, usecols=(0,1), unpack=True)
#wf Figurenachsenfrequenz in rpm
#t Zeit fr 10 Umlufe

wf = 2*np.pi*wf/60 #Umrechnen in Kreisfrequenz
w = 2*np.pi*10/t #Umrechnen in die Umlauffrequenz

dwf = 2*np.pi*2/60
dw = 20*np.pi*0.8/(t**2)

# Fit mit einer Exponentialfunktion
def fit(x, m):
    return m*x

#Fit durchfhren
popt, pcov = curve_fit(fit, wf, w, maxfev=5000, sigma=dw, p0=[0.06])
perr = np.sqrt(np.diag(pcov))
x=np.arange(0, 65, 0.1)

#Gte des Fits
chisquare=np.sum((((fit(wf,*popt))-w)/(dw))**2)
dof=9 #degrees of freedom, Freiheitsgrad Anzahl Messwerte (10) - Fitparameter (1)
chisquare_red=chisquare/dof
from scipy.stats import chi2 
prob=round(1-chi2.cdf(chisquare,dof),2)*100 


#Plotten
plt.errorbar(wf, w, xerr=dwf, yerr=dw, linestyle='None', marker='x', label='$Messwerte$')
plt.plot(x, fit(x, *popt), label='$Fit: f(x)=m*x$')
plt.xlabel('$Drehfrequenz\ \omega_{F} \ in\ [\\frac{1}{s}]$')
plt.ylabel('$Umlauffrequenz\ momentane\ Drehachse\ \Omega \ in\ [\\frac{1}{s}]$')
plt.title(r'Diagramm 3: Umlauffrequenz $\Omega(\omega_{F})$')
#plt.yscale('log')
plt.grid(True)
plt.axis((0, 63, 0, 4.0))
#plt.rcParams["figure.figsize"][0] = 8
#plt.rcParams["figure.figsize"][1] = 5
plt.rcParams["font.family"]='serif'
plt.legend(title='Messwerte mit Fit', borderpad=1.0, borderaxespad=1, loc='best', shadow='true')

#Text im Plot
plt.text(30, 1.2, u'$m_{fit}=%.5f \pm%.5f$' %(popt[0], perr[0]), fontsize=10)
plt.text(30, 0.9, u'$\chi^{2}=%.2f$' %(chisquare), fontsize=10)
plt.text(30, 0.6, u'$\chi_{red}^{2}=%.2f $' %(chisquare_red), fontsize=10)
plt.text(30, 0.3, u'$P=%.2f $' %(prob) + '%', fontsize=10)

print('Werte in roher Form:')
print('Beste Werte aus Fit:')
print(popt)
print('Fehler der Werte aus Fit (Wurzel der Kovarianzmatrixwerte)')
print(perr)
print('Chiquadratsumme:')
print (chisquare)
print("Fitwahrscheinlichkeit="+str(prob)+"%")

#plt.savefig('v213_3.pdf',format='pdf')
```

```python
#Aufgabe 5: Nutuation

#Import der Module
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

#Einlesen der Daten in folgender Reihenfolge:
wf, wn = np.loadtxt('tab5.txt', skiprows=1, usecols=(0,1), unpack=True)
#wf Figurenachsenfrequenz in rpm
#wn Nutuationsfrequenz in rpm

wf = 2*np.pi*wf/60 #Umrechnen
wn = 2*np.pi*wn/60 #Umrechnen

dwf = 2*np.pi*5/60
dwn = 2*np.pi*5/60

# Fit mit einer Exponentialfunktion
def fit(x, m):
    return m*x

#Fit durchfhren
popt, pcov = curve_fit(fit, wf, wn, maxfev=5000, sigma=dwn, p0=[0.9])
perr = np.sqrt(np.diag(pcov))
x=np.arange(0, 100, 0.1)

#Gte des Fits
chisquare=np.sum((((fit(wf,*popt))-wn)/(dwn))**2)
dof=9 #degrees of freedom, Freiheitsgrad Anzahl Messwerte (10) - Fitparameter (1)
chisquare_red=chisquare/dof
from scipy.stats import chi2 
prob=round(1-chi2.cdf(chisquare,dof),2)*100 


#Plotten
plt.errorbar(wf, wn, xerr=dwf, yerr=dwn, linestyle='None', marker='x', label='$Messwerte$')
plt.plot(x, fit(x, *popt), label='$Fit: f(x)=m*x$')
plt.xlabel('$Drehfrequenz\ \omega_{F} \ in\ [\\frac{1}{s}]$')
plt.ylabel('$Nutuationsfrequenz\ \omega_{N} \ in\ [\\frac{1}{s}]$')
plt.title(r'Diagramm 4: Nutuationsfrequenz $\omega_{N}$ ber Drehfrequenz $\omega_{F}$')
plt.grid(True)
plt.axis((0, 82, 0, 78))
#plt.rcParams["figure.figsize"][0] = 8
#plt.rcParams["figure.figsize"][1] = 5
plt.rcParams["font.family"]='serif'
plt.legend(title='Messwerte mit Fit', borderpad=1.0, borderaxespad=1, loc='best', shadow='true')

#Text im Plot
plt.text(52, 31, u'$m_{fit}=%.4f \pm%.4f$' %(popt[0], perr[0]), fontsize=10)
plt.text(52, 26, u'$\chi^{2}=%.2f$' %(chisquare), fontsize=10)
plt.text(52, 21, u'$\chi_{red}^{2}=%.2f $' %(chisquare_red), fontsize=10)
plt.text(52, 16, u'$P=%.2f $' %(prob) + '%', fontsize=10)

print('Werte in roher Form:')
print('Beste Werte aus Fit:')
print(popt)
print('Fehler der Werte aus Fit (Wurzel der Kovarianzmatrixwerte)')
print(perr)
print('Chiquadratsumme:')
print (chisquare)
print("Fitwahrscheinlichkeit="+str(prob)+"%")

#plt.savefig('v213_4.pdf',format='pdf')
```
