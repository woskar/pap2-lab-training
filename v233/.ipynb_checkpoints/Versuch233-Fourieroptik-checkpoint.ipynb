```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
#%%
# Zu Aufgabe 1

# Gereriere Ordnungszahlen
n = np.arange(1,6,1)

# Trage die Differenzen der Messwerte manuell ein
d_min = np.array([1715-1488,1824-1379,1936-1263,2045-1157,2158-1047])
# Lese Maxima-Abstnde ein
d_max = np.array([1763-1441,1878-1329,1992-1216,2094-1110,2209-996])
       
# Hier kommen andere Werte heraus als bei der Eichmessung 
# Eichung also Potenziell ungenau
# => mgliche Fehlerquelle


#Definiere linearen Fit

linear = lambda x,s,b : s*x + b
potp, pcov = curve_fit(linear, n, d_min, sigma=5.7)

perr = np.sqrt(np.diag(pcov))
x= np.arange(0,7,1)

plt.plot(x, linear(x, *potp), label='Linearer Fit der Messwerte')
plt.errorbar(n,d_min, yerr=5.7, color='k', fmt='o', label='Abstand der Beugungsminima')
 
plt.title('Diagramm 20: Abstand der Beugungsminima in Abhngigkeit \n ihrer Beugungsordnung')
plt.xlabel('Ordnung des Beugungsminima')
plt.ylabel('Abstand der Beugungsminima in [$px$]')
plt.xlim(0,6)
plt.ylim(0,1500)

# Stelle die Geradengleichung nun um, um die Abstnde der Maxima
# Eintragen zu knnen und berechne den Fehler der Ordnung

def ord_max(y):
    return (y - potp[1])/potp[0]

def d_ord_max(y):
    return np.sqrt((2.8/potp[0])**2 + (perr[1]/potp[0])**2 + ((y-potp[1])*perr[0]/(potp[0]**2))**2)
    

     
plt.scatter(ord_max(d_max), d_max, marker='x',color='red', label='Abstand der Beugungsmaxima')
plt.text(0.2,900,'Fit: $F(n)=(${:.4}$\pm${:.2}$)n+(${:.3}$\pm${:.2}$)$'.format(potp[0],perr[0],potp[1], perr[1]))
plt.legend(loc='upper left')

plt.savefig('fig1.pdf', format='PDF')

print(d_max)
print(ord_max(d_max))
print(d_ord_max(d_max))

#%%

```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# IX.1: Berechnung der Beugungsbilder 
# Definiere die Beugungsfunktion des Einfachspaltes
def beugung_spalt(x):
    return np.sinc(x)**2

# Definiere die Beugungsfunktion des Doppelspaltes
# Normiere, damit I_0 auch wirklich 1 ist
def beugung_doppelspalt(x,nu):
    return np.sinc(x)**2*np.cos(np.pi*nu*x)**2
```

```python
#Anzahl a der Nebenmaxima die dargestellt werden sollen:
a=1
#Verhltnis von Spaltabstand zu Spaltbreite:
v=2.29
#Abszissenwerte:
x=np.linspace(-(a+1),a+1,200)
#Erstellung der Graphik
plt.plot(x,beugung_spalt(x), label='Spalt')
plt.plot(x,beugung_doppelspalt(x, v), label='Doppelspalt')
plt.xlabel('$\pi x$')
plt.ylabel('Intensitt / b.E.')
plt.title('Beugung am Spalt und Doppelspalt')
plt.ylim((0,1.1))
plt.legend()
plt.text(0.55,0.8, r'Verhltnis $\nu=${:.3}'.format(v))

plt.savefig('Beugung_Spalt_Doppelspalt_Fehlersigmaminus.pdf', format="pdf")
```

```python
#Finde die Intensittsmaxima der Theoretischen Verteilung
def maxima(nu):
    y = beugung_doppelspalt(x, nu)
    y_max=[]
    for i in range(9):
        y_max.append(np.max(y[22*i:22*(i+1)]))
    #Normiere auch wirklich auf I_0    
    y_max = y_max/y_max[4]
    return y_max
print('Maxima fr nu=2.35:')
print(maxima(2.35))
print('Maxima fr nu=2.41:')
print(maxima(2.41))
print('Maxima fr nu=2.29:')
print(maxima(2.29))
```

```python
#IX.2 Berechnung der modifizierten Spaltbilder

#Fr den Einfachspalt:
#Integrand
def spalt(k):
    return d/np.pi*np.sin(k*d/2)/(k*d/2)*np.cos(y*k)

d=0.428 #Spaltbreite auf 1 setzen
n1=10 #Integrationsbereich

#y-Werte
Y1=np.linspace(-d,d,500*d)
```

```python
#Berechnung des resultierenden Objektbilds am Spalt
from scipy.integrate import quad
f_modifiziert1=[]
for i in range(len(Y1)):
    y=Y1[i]
    result1, error1= quad(spalt, -2*np.pi*n1/d, 0)
    f_modifiziert1.append(result1**2)

#Kurve auf 1 normieren
#Behalte normierte Version im Verhltnis zu n1=1 mit
norm1=1.38999
norm= np.max(f_modifiziert1)/norm1
f_modifiziert1=f_modifiziert1/np.max(f_modifiziert1)

#Darstellung des modifizierten Spaltbilds
plt.plot(Y1/d, f_modifiziert1)
plt.xlabel('y / d')
plt.ylabel('Intensitt / b.E.')
plt.title('Modifiziertes Objektbild am Spalt')
plt.ylim((0,1.1))
plt.text(-0.9,0.05,'Integration bis zum \n{}-ten Minimum'.format(n1))

plt.savefig('spalt6.pdf', format="pdf")
```

```python
#Finde die Maxima der Kurve
max_mask = np.r_[f_modifiziert1[1:] < f_modifiziert1[:-1], True] & np.r_[True,f_modifiziert1[:-1] < f_modifiziert1[1:]];
f_max = f_modifiziert1[max_mask]
Y_max = Y1[max_mask]
Y_max = Y_max[f_max>10**(-2)]
f_max = f_max[f_max>10**(-2)]


#Finde die Minima der Kurve
min_mask = np.r_[f_modifiziert1[1:] > f_modifiziert1[:-1], True] & np.r_[True,f_modifiziert1[:-1] > f_modifiziert1[1:]];
f_min = f_modifiziert1[min_mask]
Y_min = Y1[min_mask]
Y_min = Y_min[f_min>10**(-2)]
f_min = f_min[f_min>10**(-2)]

# Werte Ergebnisse aus
# Mittelwerte ber die Intensitten
int_max= np.mean(f_max)*norm
d_int_max = np.std(f_max)*norm
int_min= np.mean(f_min)*norm
d_int_min= np.std(f_min)*norm

print('Mittelwert ber die Intensitten:')
print('Maxima: {} +- {:.2}'.format(int_max, d_int_max))
print('Minima: {} +- {:.2}'.format(int_min, d_int_min))

# Abstnde zwischen den Extrema:
dy_max=[]
dy_min=[]
for i in range(n1-1):
    dy_max.append(Y_max[i+1]-Y_max[i])
    
for i in range(n1-2):
    dy_min.append(Y_min[i]-Y_min[i-1])
print('Abstnde Maxima:')
print(dy_max)
print('Abstnde Minima:')
print(dy_min)
```
